```python
# importing the libraries for data loading,manupulating and plotting
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading the datasets
customers_path = "Dataset/Customers.csv"
transactions_path ="Dataset/Transactions.csv"
products_path ="Dataset/Products.csv"

#reading the datasets
customers = pd.read_csv(customers_path)
transactions=pd.read_csv(transactions_path)
products=pd.read_csv(products_path)
```

# 1.Customer Aquisition Overtime

```python
# Convert 'SignupDate' to datetime
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'], errors='coerce')

# Extract the year and month of signup from 'SignupDate'
customers['SignupMonth'] = customers['SignupDate'].dt.to_period('M')

# we Count customers acquired each month
customer_acquisition = customers.groupby('SignupMonth').size().cumsum()

# Plotss
plt.figure(figsize=(12, 6))
customer_acquisition.plot(kind='line', marker='o', title='Customer Acquisition Over Time')
plt.xlabel('Signup Month')
plt.ylabel('Cumulative Customers')
plt.grid(True)
plt.show()
```

# 2.Average Revenue Per Customer by Region

```python
merged_data = transactions.merge(customers, on='CustomerID')

# Calculate average revenue per customer by region
region_avg_revenue = merged_data.groupby('Region').agg(
    TotalRevenue=('TotalValue', 'sum'),
    CustomerCount=('CustomerID', 'nunique')
).reset_index()

# Calculate average revenue per customer
region_avg_revenue['AvgRevenuePerCustomer'] = region_avg_revenue['TotalRevenue'] / region_avg_revenue['CustomerCount']

# Plot average revenue per customer as a bar chart
plt.figure(figsize=(10, 6))
plt.bar(region_avg_revenue['Region'], region_avg_revenue['AvgRevenuePerCustomer'], color='lightcoral')
plt.title('Average Revenue Per Customer by Region', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Average Revenue Per Customer ($)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 3.Top 10 Products by Revenue and Quantity Sold

```python
# Calculate total revenue and quantity sold for each product
product_performance = transactions.groupby('ProductID').agg(
    TotalRevenue=('TotalValue', 'sum'),
    TotalQuantity=('Quantity', 'sum')
).reset_index()

# Merge with product details for better context
product_performance = product_performance.merge(products[['ProductID', 'ProductName']], on='ProductID')

# Sort products by total revenue and pick top 10
top_products = product_performance.sort_values(by='TotalRevenue', ascending=False).head(10)

# Plot the dual-axis chart
fig, ax1 = plt.subplots(figsize=(14, 8))

# Bar chart for Total Revenue
ax1.bar(top_products['ProductName'], top_products['TotalRevenue'], color='skyblue')
ax1.set_xlabel('Product Name', fontsize=12)
ax1.set_ylabel('Total Revenue ($)', fontsize=12, color='blue')
ax1.set_xticklabels(top_products['ProductName'], rotation=45, ha='right')

# Line chart for Quantity Sold (secondary axis)
ax2 = ax1.twinx()
ax2.plot(top_products['ProductName'], top_products['TotalQuantity'], color='orange', marker='o')
ax2.set_ylabel('Total Quantity Sold', fontsize=12, color='orange')

# Add Legends and Display
fig.tight_layout()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

# 4.Seasonal Trends in Transactions

```python
# Convert 'TransactionDate' to datetime and extract month
transactions['TransactionMonth'] = pd.to_datetime(transactions['TransactionDate']).dt.to_period('M')

# Count transactions per month
monthly_transactions = transactions.groupby('TransactionMonth').size().reset_index(name='TransactionCount')

# Plot
plt.figure(figsize=(12, 6))
plt.plot(monthly_transactions['TransactionMonth'].astype(str), monthly_transactions['TransactionCount'], marker='o', color='teal')
plt.title('Seasonal Trends in Transactions', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Summary-Revenue vs Customer Engagement by Region

```python
merged_data = transactions.merge(customers[['CustomerID', 'Region']], on='CustomerID')

# Calculate total revenue and customer count by region
region_data = merged_data.groupby('Region').agg(
    TotalRevenue=('TotalValue', 'sum'),
    CustomerCount=('CustomerID', 'nunique')
).reset_index()

# Plot
fig, ax1 = plt.subplots(figsize=(12, 6))

# Revenue bar chart
ax1.bar(region_data['Region'], region_data['TotalRevenue'], color='lightblue')
ax1.set_xlabel('Region')
ax1.set_ylabel('Total Revenue ($)')
plt.xticks(rotation=45)

# Customer count on secondary axis
ax2 = ax1.twinx()
ax2.bar(region_data['Region'], region_data['CustomerCount'], width=0.4, color='orange')
ax2.set_ylabel('Number of Customers')

# Show plot
fig.tight_layout()
plt.show()
```
